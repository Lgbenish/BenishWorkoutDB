```python
import pandas as pd
import openpyxl 
import pyodbc
```

```python
# Connection string for Windows Auth
conn_str = (
    "DRIVER={SQL Server};"
    "SERVER=;"
    "DATABASE=BenishWorkoutDB;"
    "Trusted_Connection=yes;"
)

# Test connection
try:
    conn = pyodbc.connect(conn_str)
    cursor = conn.cursor()
    print("✅ Connected to SQL Server!")
except Exception as e:
    print("❌ Connection failed:", e)
```

```python
# Your file path and sheet
WorkoutFilePath = r"
sheet_name = "Raw"

# Load the Excel file
df = pd.read_excel(WorkoutFilePath, sheet_name=sheet_name, engine='openpyxl')

# Preview the first few rows
df.head()
```

```python
first_row = df.iloc[0]
for col in df.columns:
    print(f"{col}: {type(first_row[col])} - {first_row[col]}")
```

```python
df_workouts = df[df['WorkoutFLG'] == 1].copy()

pd.set_option('display.max_columns', None)  
pd.set_option('display.max_colwidth', 10000)
df_workouts.describe()
```

```python
for index, row in df_workouts.iterrows():
    try:
        # === Construct pseudo-key ===
        workout_date = pd.to_datetime(row['Date']).strftime('%Y-%m-%d')
        location = str(row['LocationNM']).strip().replace(" ", "_")
        workout_flg = str(row['WorkoutFLG']).strip()
        individual = str(row.get('WorkoutIndividual', 'Unknown')).strip().replace(" ", "_")
        workout_nm = str(row['WorkoutNM']).strip().replace(" ", "_")

        workout_key = f"{workout_date}_{location}_{workout_flg}_{individual}_{workout_nm}"

        # === Check for existing workout ===
        cursor.execute("SELECT WorkoutID FROM Workouts WHERE WorkoutKey = ?", (workout_key,))
        existing = cursor.fetchone()
        if existing:
            print(f"⏩ Skipping existing workout: {workout_key}")
            continue

        # === Insert new workout ===
        cursor.execute("""
            INSERT INTO Workouts (
                WorkoutDate, MonthNM, WeekOfYearNBR, WeekdayNM, YearNBR,
                WeekdayFLG, WeekendFLG, DayDenominatorNBR, HistoricalFLG,
                BodyWeightNBR, LocationNM, NotesDSC, WorkoutFLG,
                WorkoutGrouperDSC, WorkoutNM, TotalVolume, WorkoutIndividual, WorkoutKey
            )
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (
            workout_date,
            row['MonthNM'], row['WeekOfYearNBR'], row['WeekdayNM'], row['YearNBR'],
            row['WeekdayFLG'], row['WeekendFLG'], row['DayDenominatorNBR'], row['HistoricalFLG'],
            row['BodyWeightNBR'], row['LocationNM'], row['NotesDSC'], workout_flg,
            row['WorkoutGrouperDSC'], row['WorkoutNM'], row['TotalVolume'], individual, workout_key
        ))

        # === Retrieve new WorkoutID ===
        cursor.execute("SELECT SCOPE_IDENTITY()")
        workout_id = cursor.fetchone()[0]

        print(f"✅ Inserted workout {workout_id} ({workout_key})")

    except Exception as e:
        print(f"❌ Error on row {index}: {e}")
        continue

conn.commit()
```
